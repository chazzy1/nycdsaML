```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
from scipy.stats import norm, skew
from scipy import stats
import numpy as np
import seaborn as sns

sns.set()
```

```python
#Data loading
train_set = pd.read_csv('./data/train.csv')
test_set = pd.read_csv('./data/test.csv')
```

```python
#Outlier

tmpDf = train_set.copy(deep=True)

numeric_columns = tmpDf.dtypes[tmpDf.dtypes != "object"].index
for col in numeric_columns:
    tmpDf[col].fillna(0, inplace=True)
    
z = np.abs(stats.zscore(tmpDf[numeric_columns]))



row, col = np.where(z > 5)

df = pd.DataFrame({"row":row, "col":col})
columns_with_outliers = {}
for i in zip(row,col):
    row, col = i
    col_name = tmpDf[numeric_columns].columns[col]
    columns_with_outliers.setdefault(col_name,[]).append(row)
    
print(columns_with_outliers)


rows_count = df.groupby(['row']).count()
rows_count[rows_count.col>2]

#plt.bar(rows_count.index, rows_count.col, width=10)
```
